```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/air_pollution_indicators_lka.csv")
print(dataset)
```

```python
dataset.head()
```

```python
#Removing the row at index 0 as it doesn't contribute to any insights and resetting the index
dataset = dataset.iloc[1:].reset_index(drop=True)
```

```python
#finding the number of duplicated entries for each attribute in the dataset
duplicated_values = dataset.duplicated().sum()
print("These are the total count of duplicates in the dataset: ")
print("\n", duplicated_values)
```

```python
dataset.dtypes
```

```python
#changing the data types
category_columns = [
    "GHO (CODE)", "GHO (DISPLAY)", "REGION (CODE)", "REGION (DISPLAY)",
    "COUNTRY (CODE)", "COUNTRY (DISPLAY)",
    "DIMENSION (TYPE)", "DIMENSION (CODE)", "DIMENSION (NAME)"
]

for column in category_columns:
    dataset[column] = dataset[column].astype("category")

string_columns = ["YEAR (DISPLAY)","Value"]

for column in string_columns:
  dataset[column]=dataset[column].astype("string")

integer_columns = ['STARTYEAR','ENDYEAR']

for column in integer_columns:
    dataset[column] = pd.to_numeric(dataset[column], errors='coerce')

    dataset[column] = dataset[column].astype("Int64")

float_columns = ["Numeric", 'Low','High']

for column in float_columns:
    dataset[column] = pd.to_numeric(dataset[column], errors='coerce')
    dataset[column] = dataset[column].astype("float64")


print(dataset.dtypes)
```

```python
dataset.dtypes
```

```python
#Finding missing values in the dataset
dataset.isnull().sum()
```

```python
#Outlier detection for each of the attributes
plt.figure(figsize=(10, 6))
#Using seaborn to illustrate a boxplot for the following attributes
sns.boxplot(data=dataset[['Numeric', 'Low','High']])
plt.title("Boxplot for Sales Outlier Detection")
plt.show()
```

```python
columns = ['Numeric', 'Low','High' ,'STARTYEAR','ENDYEAR']


for column in columns:
    data = dataset[column]

    # Create histogram
    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=30, density=True, alpha=0.6, color='darkgreen', edgecolor='black', label='Actual Data')

    # Generate a smoothed line using Kernel Density Estimation (KDE)
    sns.kdeplot(data, color='red', linewidth=2, label='Smoothed KDE')

    # Labels and Title
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.title(f'Histogram with Smoothed Curve - {column}')
    plt.legend()

    # Show plot
    plt.show()
```

```python
#Handling Missing values in the dataset with median imputation
dataset["Low"] = dataset["Low"].fillna(dataset["Low"].median())
dataset['High'] = dataset['High'].fillna(dataset['High'].median())
```

```python
#Ensuring there are no missing values in the dataset after imputation has been implemented
dataset.isnull().sum()
```
